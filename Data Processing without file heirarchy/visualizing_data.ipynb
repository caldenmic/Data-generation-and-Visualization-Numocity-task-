```python
from plot_functions import *
import pandas as pd
import numpy as np

df_charge_point_and_station = pd.read_csv('charge_point_and_station_summary.csv')
df_charge_point = pd.read_csv('charge_point_summary.csv')
df_charge_station = pd.read_csv('charge_station_summary.csv')

line_plot(df_charge_point.groupby(['date', 'txn.chargePointId']).agg({'Total_amount_charged': 'sum'}).reset_index(), 'date', 'Total_amount_charged', rotation=75, figsize=(14,10), hue='txn.chargePointId')
```

```python
df_weekend_and_weekday = df_charge_point.groupby(['day_of_the_week']).agg({'Total_amount_charged': 'sum', 'txn.deliveredWh': 'sum'}).reset_index()
weekday_data = df_weekend_and_weekday[df_weekend_and_weekday['day_of_the_week'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])]
weekend_data = df_weekend_and_weekday[df_weekend_and_weekday['day_of_the_week'].isin(['Sunday', 'Saturday'])]
```
